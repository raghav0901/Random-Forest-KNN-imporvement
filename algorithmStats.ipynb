```python
#Comparing With and Without KNN improvement
import matplotlib.pyplot as plt
x = [1,5,10]
woutImpOrKNN = [87.536,94.029,93.217]
wImpANDKNN = [89.333,95.420,94.957]
plt.plot(x,woutImpOrKNN, marker='D', label = 'RegRF')
plt.plot(x,wImpANDKNN,  marker='D', label = 'RF+KNN')
plt.title("Accuracy-TreeSize graph For DataSet 1")
plt.xlabel('Tree Size')
plt.ylabel('Accuracy')
plt.legend(['RegularRF', 'RF+KNN_Imp'])
plt.show()
```
